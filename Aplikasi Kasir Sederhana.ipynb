```python
import matplotlib.pyplot as plt

# ==============================
# DATA LOGIN
# ==============================
USERNAME = "admin"
PASSWORD = "12345"

# ==============================
# DATA BARANG
# ==============================
barang = {
    1: {"nama": "Baju", "harga": 120000},
    2: {"nama": "Celana", "harga": 90000},
    3: {"nama": "Sepatu", "harga": 250000},
    4: {"nama": "Topi", "harga": 50000}
}

# ==============================
# DATA PENJUALAN (untuk grafik)
# ==============================
data_penjualan = {}

# ==============================
# FUNGSI LOGIN
# ==============================
def login():
    print("=== LOGIN SISTEM KASIR ===")
    while True:
        user = input("Username : ")
        pw = input("Password : ")

        if user == USERNAME and pw == PASSWORD:
            print("Login berhasil!\n")
            break
        else:
            print("Login gagal, coba lagi!\n")

# ==============================
# TAMPILKAN LIST BARANG
# ==============================
def tampil_barang():
    print("\n=== LIST BARANG ===")
    for kode, item in barang.items():
        print(f"{kode}. {item['nama']} - Rp {item['harga']}")
    print()

# ==============================
# INPUT PESANAN
# ==============================
def input_pesanan():
    nama_konsumen = input("\nNama Konsumen : ")
    pesanan = []
    total = 0

    while True:
        tampil_barang()
        kode = int(input("Pilih kode barang (0 untuk selesai): "))

        if kode == 0:
            break

        if kode in barang:
            jumlah = int(input("Jumlah : "))
            subtotal = barang[kode]["harga"] * jumlah

            pesanan.append((barang[kode]["nama"], jumlah, subtotal))
            total += subtotal
        else:
            print("Kode tidak valid!")

    print("\nTotal Belanja : Rp", total)

    # ==============================
    # PEMBAYARAN
    # ==============================
    bayar = int(input("Masukkan pembayaran : Rp "))

    if bayar < total:
        print("Pembayaran kurang!")
        return

    kembalian = bayar - total

    # ==============================
    # SIMPAN DATA PENJUALAN
    # ==============================
    if nama_konsumen in data_penjualan:
        data_penjualan[nama_konsumen] += total
    else:
        data_penjualan[nama_konsumen] = total

    # ==============================
    # CETAK STRUK
    # ==============================
    print("\n===== STRUK PEMBELIAN =====")
    print("Konsumen :", nama_konsumen)

    for item in pesanan:
        print(f"{item[0]} x{item[1]} = Rp {item[2]}")

    print("---------------------------")
    print("Total     : Rp", total)
    print("Bayar     : Rp", bayar)
    print("Kembalian : Rp", kembalian)
    print("===========================\n")

# ==============================
# GRAFIK PENJUALAN
# ==============================
def tampil_grafik():
    if not data_penjualan:
        print("Belum ada data penjualan!\n")
        return

    nama = list(data_penjualan.keys())
    total = list(data_penjualan.values())

    plt.bar(nama, total)
    plt.title("Grafik Penjualan per Konsumen")
    plt.xlabel("Konsumen")
    plt.ylabel("Total Belanja")
    plt.show()

# ==============================
# MENU UTAMA
# ==============================
def menu():
    while True:
        print("=== MENU UTAMA ===")
        print("1. Tampil List Barang")
        print("2. Input Pesanan")
        print("3. Tampil Grafik Penjualan")
        print("4. Keluar")

        pilih = input("Pilih menu : ")

        if pilih == "1":
            tampil_barang()

        elif pilih == "2":
            input_pesanan()

        elif pilih == "3":
            tampil_grafik()

        elif pilih == "4":
            print("Terima kasih!")
            break

        else:
            print("Menu tidak tersedia!\n")

# ==============================
# PROGRAM UTAMA
# ==============================
login()
menu()1
```
