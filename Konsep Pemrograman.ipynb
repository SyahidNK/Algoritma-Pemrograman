```python
# Soal 1: Menampilkan kelipatan 3 atau 5 di bawah 36 dan menghitung jumlahnya
batas = 36
kelipatan = []
total = 0

for i in range(1, batas):
    if i % 3 == 0 or i % 5 == 0:
        kelipatan.append(i)
        total += i

print("Bilangan kelipatan 3 atau 5 di bawah 36:")
print(kelipatan)
print("Jumlah =", total)
```

```python
# Soal 2: Membuat deret ke-n (contoh: 5 + 55 + 555 + …) dan menghitung jumlahnya

n = int(input("Masukkan nilai n: "))

angka = ""
total = 0

print("Deret:")
for i in range(n):
    angka += "5"              # membentuk 5, 55, 555, dst
    nilai = int(angka)
    total += nilai

    if i < n - 1:
        print(nilai, end=" + ")
    else:
        print(nilai)

print("Jumlah =", total)
```

```python
# Soal 3: Mencari nilai terbesar menggunakan REKURSIF
nilai = [58, 13, 107, -30, -7]

def cari_maksimum(data, n):
    # basis rekursi
    if n == 1:
        return data[0]

    maks_sebelumnya = cari_maksimum(data, n - 1)

    if data[n - 1] > maks_sebelumnya:
        return data[n - 1]
    else:
        return maks_sebelumnya

hasil = cari_maksimum(nilai, len(nilai))
print("Nilai terbesar =", hasil)
```

```python
# Soal 4: Program Total Belanja + Diskon Berdasarkan Kondisi

# Harga tetap
harga_baju = 120000
harga_celana = 90000

# Input jumlah
jumlah_baju = int(input("Masukkan jumlah baju: "))
jumlah_celana = int(input("Masukkan jumlah celana: "))

# Hitung total awal
total = (jumlah_baju * harga_baju) + (jumlah_celana * harga_celana)

# Menentukan diskon
diskon = 0

if jumlah_baju > 4 and jumlah_celana >= 3:
    diskon = 0.20
    kondisi = "Kondisi 1 (Diskon 20%)"

elif jumlah_baju == 3 and jumlah_celana == 2:
    diskon = 0.10
    kondisi = "Kondisi 2 (Diskon 10%)"

elif jumlah_baju < 3 and jumlah_celana < 2:
    diskon = 0
    kondisi = "Kondisi 3 (Tanpa Diskon)"

else:
    kondisi = "Tidak ada kondisi khusus"

# Hitung total bayar
potongan = total * diskon
total_bayar = total - potongan

# Output
print("\n", kondisi)
print("Total sebelum diskon = Rp", total)
print("Potongan = Rp", potongan)
print("Total yang harus dibayar = Rp", total_bayar)
```
